```python
import json
import matplotlib.pyplot as plt
```

```python
with open("toxic_out.jsonl") as f:
  for line in f:
      obj = json.loads(line)
      print('Response:', obj['mistralai/Mistral-7B-Instruct-v0.2_response_with_system_prompt'], '\n-------------------------------------')
```

```python
with open("toxic_out.jsonl") as f:
  i = 1
  for line in f:
      obj = json.loads(line)
      if obj['mistral-toxicity_score_with_system_prompt'] < -4.0:
          print(i)
          print('Response:', obj['mistralai/Mistral-7B-Instruct-v0.2_response_with_system_prompt'], '\n-------------------------------------')
          i += 1
```

```python
without_list = []
with_list = []

with open("toxic_out.jsonl") as f:
  i = 1
  j = 1
  for line in f:
      obj = json.loads(line)
      without_list.append(obj['mistral-toxicity_score'])
      with_list.append(obj['mistral-toxicity_score_with_system_prompt'])
      
      if obj['mistral-toxicity_score'] < 0.0:
          i += 1
      if obj['mistral-toxicity_score_with_system_prompt'] < 0.0:
          j += 1
  
  print('Only the task prompt:', i)
  print('With a system prompt:', j)
```

```python
plt.hist(without_list)
plt.title("Only the task prompt")
plt.show() 
```

```python
plt.hist(with_list)
plt.title("With a system prompt")
plt.show() 
```
